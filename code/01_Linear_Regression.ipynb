# Linear Regression Demo

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

用 numpy 亂數產生 100 個隨機點

```python
x_data = np.random.rand(100).astype(np.float32)
y_data = x_data * 0.1 + 0.3
```

初始化 權重直 W 和 偏差值 b，並建立線性方程 y = W * x + b

```python
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.zeros([1]))
y = W * x_data + b
```

定義 loss function 為 mean square errors(均方誤差)

```python
loss = tf.reduce_mean(tf.square(y - y_data))
```

定義 optimize function 為 grandient descent(梯度下降)

```python
optimizer = tf.train.GradientDescentOptimizer(0.2)
```

Minimize loss function

```python
train = optimizer.minimize(loss)
```

初始化變數

```python
init = tf.global_variables_initializer()
```

啟動 Graph

```python
sess = tf.Session()
sess.run(init)
```

資料點擬和

```python
itera = 200
for step in range(itera):
    sess.run(train)
    
    if step % 20 == 0:
        print('第{0}次訓練: 權重值: {1}, 偏移值: {2}'.format(step, sess.run(W), sess.run(b)))
        plt.plot(x_data, y_data, 'ro', label = 'Original data')
        plt.plot(x_data, sess.run(W) * x_data + sess.run(b), 'b', label = 'Fitted line')
        plt.legend()
        plt.show()
```
